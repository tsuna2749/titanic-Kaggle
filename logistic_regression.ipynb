```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
train_set = pd.read_csv('train.csv')
train_set
```

```python
train_set = train_set[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch']]
train_set
```

```python
train_set = train_set.dropna(axis = 0).reset_index()
train_set
```

```python
for i in range(len(train_set)):
    if train_set.loc[i, 'Sex'] == 'male':
        train_set.loc[i, 'Sex'] = 1
    else:
        train_set.loc[i, 'Sex'] = 0
```

```python
train_set
```

```python
# devide the train set into input X and output Y
X_train = train_set[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch']]
Y_train = train_set[['Survived']]
```

```python
X_train
```

```python
Y_train
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
```

```python
len(X_train)
```

```python
# Fitting Logistic Regression to the training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, Y_train)
```

```python
classifier
```

```python
# import the test set
test_set = pd.read_csv('test.csv')
test_set
```

```python
user_id = test_set[['PassengerId']]
test_set = test_set[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch']]
user_id
```

```python
for i in range(len(test_set)):
    if test_set.loc[i, 'Sex'] == 'male':
        test_set.loc[i, 'Sex'] = 1
    else:
        test_set.loc[i, 'Sex'] = 0
```

```python
test_set
```

```python
test_set.isnull().any()
```

```python
# fill nan in age with the mean of the column
age_avg = test_set['Age'].sum() / len(test_set)
age_avg = round(age_avg, 2)
age_avg
```

```python
test_set = test_set.fillna(value = age_avg)
test_set.isnull().any()
```

```python
test_set
```

```python
# predict the result of test_set using the classifier we get
test_set = sc_X.fit_transform(test_set)
pred_res = classifier.predict(test_set)
```

```python
pred_res
```

```python
user_id['Survived'] = pred_res
```

```python
user_id
```

```python
# output the prediction
user_id.to_csv('logistic_regression.csv', index = False)
```
